## Aufgabe 15: Lösungsvorschlag

```julia
using Plots
```

```julia
instab(x) = (1-cos(x))/x^2;
```

Instabilitäten in der Nähe von $x=0$.

```julia
x = 10.0f0 .^ LinRange{Float32}(-4, -2, 200)
scatter(x, instab.(x), markershape=:+, xscale=:log10)
```

Verwendung von trigonometrischen Formeln für neuen Ausdruck:

```julia
f1(x) = (2*sin(x/2)^2)/(x^2)
```

```julia
x = 10.0f0 .^ LinRange{Float32}(-4, -2, 200)
scatter(x, instab.(x), markershape=:+, xscale=:log10)
scatter!(x, f1.(x), markershape=:+)
```

```julia
x = 1.2345678f-2
instab(x), f1(x)
```

Verwendung des Kahan-Tricks:

```julia
f2(x) = let w = cos(x)
    w == 1 ? 0.5 : (1-w)/acos(w)^2
end
```

```julia
x = 10.0f0 .^ LinRange{Float32}(-4, 0, 200)
scatter(x, instab.(x), markershape=:+, xscale=:log10)
scatter!(x, f1.(x), markershape=:+)
scatter!(x, f2.(x), markershape=:+)
```

```julia
x = 1.2345678f-2
instab(x), f1(x), f2(x)
```
